```python
def noise(delta, n_samples):
    return np.sqrt(2*np.log(2/delta)/n_samples)
```

```python
delta_list = np.arange(1e-3, 1, 1e-3)
sample_list = np.arange(1, 100)
x_list = []
y_list = []
z_list = []

for delta in delta_list:
    for sample in sample_list:
        x_list.append(delta)
        y_list.append(sample)
        z_list.append(noise(delta, sample))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

print("n_points: {}".format(len(x_list)))

ax.scatter(x_list, y_list, z_list)
plt.show()
```
