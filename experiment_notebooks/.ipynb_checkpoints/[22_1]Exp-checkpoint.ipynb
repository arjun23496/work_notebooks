```python
result_location = "/media/arjun/Shared/chaos/output_files_v2"
experiment_id = 6822382

directory = os.path.join(result_location, "rademacher_test_{}".format(experiment_id))
```

```python
file_list = os.listdir(directory)
file_list = [ '.'.join(file.split('.')[:2]) for file in file_list if 'params.pkl' in file ]
print(file_list)
```

```python
file_list = os.listdir(directory)
# print(file_list)

hidden_size_list = np.sort(np.unique([ int(file_name.split('_')[1]) for file_name in file_list ]))
# print(hidden_size_list)
param_list = [ compute_parameters.get_count(100, 100, [ val, val ], ['lstm', 'fcn']) 
              for val in hidden_size_list ]

startp_list = np.sort(np.unique([ file_name.split('_')[2].split('.')[0] for file_name in file_list ]))

program_list = [ int(file_name.split('_')[0]) for file_name in file_list ]

print(param_list)
print(startp_list)
```

```python
estimated_info = {}
estimated_mean = {}

for program in file_list:
    print("Program: {}".format(program))
    
    accuracy_list = []
    ce_list = []
    ce_bound_list = []
    ce_rad_list = []
    
    
    program_file_list = [ file_name[] ]
    
    for hidden_size in hidden_size_list:
        with open(os.path.join(directory, "{}.accuracy.pkl".format(program)), 'rb') as f:
            accuracy_list = pkl.load(f)

        with open(os.path.join(directory, "{}.ce.pkl".format(program)), 'rb') as f:
            ce_list = pkl.load(f)

        with open(os.path.join(directory, "{}.ceBound.pkl".format(program)), 'rb') as f:
            ce_bound_list = pkl.load(f)

        with open(os.path.join(directory, "{}.ceRad.pkl".format(program)), 'rb') as f:
            ce_rad_list = pkl.load(f)

    data_ce = []
    data_ce_bound = []
    data_min_ce_bound = []
    data_min_rad = []
    data_ce_rad = []
    data_x = []
    
    for idx, key in enumerate(range(len(param_list))):
        loss_array = ce_list[key].flatten()
        data_ce.extend(loss_array.tolist())
        data_x.extend([param_list[key]]*loss_array.shape[0])
        
        loss_array = ce_bound_list[key][0].flatten()
#         print(ce_bound_list[key])
        data_ce_bound.extend(loss_array)
        
        loss_array = ce_rad_list[key].flatten()
#         print(loss_array)
#         print(data_ce_rad)
        data_ce_rad.append(loss_array[0])
        
    data_min_ce_bound = [ np.min(data_ce_bound[:val]) for val in range(1, len(data_ce_bound)+1) ]
    data_min_rad = [ np.min(data_ce_rad[:val]) for val in range(1, len(data_ce_rad)+1) ]
    
    parameter_min = 0
    bound_min = 0
    
    estimated_info['-'.join(program.split('-')[:3])] = np.min(data_min_ce_bound)
    estimated_mean['-'.join(program.split('-')[:3])] = np.min(data_ce)
    
    prev_val = 0
    for idx, val in enumerate(data_min_ce_bound):
        if prev_val != 0:
            improv = (prev_val-val)
            if improv > 5e-4:
                parameter_min = param_list[idx]
                bound_min = val
            print("param: {}, Improvement: {}, val: {}".format(param_list[idx], improv, val))
        prev_val = val
    
    param_limit = 1e4
    
#     plt.xlabel("number of parameters")
#     plt.ylabel("accuracy")
# #     plt.scatter(data_x, data_ce, label="estimate")
#     plt.plot(param_list, accuracy_list)
# #     plt.ylim((0.06, 0.065))
#     plt.xlim((0, param_limit))
#     plt.title("Accuracy")
#     plt.show()
    
    print("Minimum parameter: {}, bound at min: {}".format(parameter_min, bound_min))
    
    plt.xlabel("number of parameters")
    plt.ylabel("cross entropy")
#     plt.scatter(data_x, data_ce, label="estimate")
#     print(data_ce_bound)
    print(param_list)
    plt.scatter(param_list, data_ce_bound, label="loss bound")
    plt.plot(param_list, data_min_ce_bound, label="minimum loss bound")
#     plt.ylim((0.06, 0.065))
    plt.xlim((0, param_limit))
    plt.title("bound")
    plt.legend()
    plt.show()
```
