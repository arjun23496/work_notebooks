```python
## Trace lengths

import pickle as pkl
import numpy as np
import pandas as pd

from misc.database import Database

import os
import re

from bokeh.io import output_notebook, show
from bokeh.layouts import gridplot
from bokeh.plotting import figure
from bokeh.palettes import Inferno, all_palettes
from bokeh.models import (CustomJS, Slider, ColumnDataSource, Legend, 
                         BasicTicker, ColorBar, ColumnDataSource,
                          LinearColorMapper, LogColorMapper,
                          PrintfTickFormatter)
from bokeh.models import Whisker, HoverTool, Span, ColorBar
from bokeh.transform import linear_cmap, log_cmap, factor_cmap, transform

from prediction_model.utils import compute_parameters

output_notebook()
```

```python
data_dir = "/media/arjun/SSD/chaos/data"
compression_db = "/media/arjun/SSD/chaos/lossy_compression_moss/compression.sql3"

db = Database()
db.open(compression_db)

files_list = os.listdir(data_dir)

ratios = []
lengths = []

for file in files_list:
    if '.pkl' in file:
#         print(file)
        with open(os.path.join(data_dir, file), 'rb') as f:
            trace = pkl.load(f)
        
        try:
            m = re.match(r"(.*)-(.*)-(.*)-(.*)-l(.*)-p(.*)-w(.*)i.analyzed-1.pkl", file)

            pname = m.group(1)
            pin_type = m.group(2)
            jvm = m.group(3)
            idata = m.group(4)
            line_size = m.group(5)
            p_param = m.group(6)
            win = m.group(7)

            mseries_name = "{}.{}.{}.{}.{}.{}.{}.4096.0".format(pname,
                                                               jvm,
                                                               pin_type,
                                                               pname,
                                                               idata,
                                                               line_size,
                                                               10000)
#             print(mseries_name)
            db.query("SELECT * FROM RATIOS WHERE mseries = \"{}\"".format(mseries_name))
    #         print(mseries_name)
    #         print(m.group(1))
            rows = db.cursor.fetchall()
            
            count = 0
            for row in rows:
                count += 1
                mseries, ratio = row
                ratios.append(ratio)
                lengths.append(trace.shape[0])
            
#             print("{} rows found".format(count))
        except AttributeError:
            pass
#         print(file)
        print("filename: {}, size: {}, ratio: {}".format(file, trace.shape, ratio))
```
